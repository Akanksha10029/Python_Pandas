```python
#Reshape pandas dataframe using melt
import pandas as pd
df=pd.read_csv('/content/Melt_weather.csv')
df
```

```python
melted=pd.melt(df,id_vars=["day"])
melted
```

```python
melted=pd.melt(df,id_vars=["day"],var_name='city',value_name="temperature")
melted
```
