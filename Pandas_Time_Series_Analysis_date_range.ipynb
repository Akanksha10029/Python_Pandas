```python
# Pandas Time Series Analysis: date_range
import pandas as pd
df= pd.read_csv("/content/date_range_aapl_no_dates.csv")
df.head()
```

```python
'''Here we are adding dates by using date_range and for we are giving starting and
ending date with the frequency of B ,i.e, Business Date which means in dates it will exclude weekends '''

rng = pd.date_range(start="6/1/2016",end="6/30/2016",freq='B')
rng
```

```python
df.set_index(rng, inplace=True)
df.head()
```

```python
df.index
```

```python
# Finding missing dates from datetimeindex
'''Here we are using D as frequency for finding weekends as well in dates'''
daily_index = pd.date_range(start="6/1/2016",end="6/30/2016",freq='D')
daily_index
```

```python
df.head(10)
```

```python
daily_index.difference(df.index)
```

```python
# Benefits of having DatetimeIndex
'''We can plot graph for analysing something according to our requirements on the basis of dates'''
import matplotlib.pyplot as plt
plt.xlabel("dates")
plt.ylabel("Open")
df.Open.plot()
```

```python
'''Here asfreq allows us to regenerate our dataframe acc. to the new frequency so we are including days as D and
method we are using is padding as pad(pad / ffill: propagate last valid observation forward to next valid
,i.e, puts the first previous value before saturday and sunday in these two)'''
df.asfreq('D',method='pad').head(10)
```

```python
df.asfreq('W',method='pad') # W stands for Weekly
```

```python
df.asfreq('H',method='pad').head(20) #H is for hourly
```

```python
# generating DatetimeIndex with periods argument
'''if we haven't specified the end date then we can use periods like till which number
we want to add date for eg. here is 72 , we are giving 72 as period which will give the dates up to 72th date'''
rng = pd.date_range('1/1/2011', periods=72, freq='W')
rng
```

```python
'''if we want to generate random numbers between any range acc. to our wish,
for eg here (0,10)i.e 0 to 10 which is 10 random numbers throughout the length range(len(rng)) i.e for 72 days'''
import numpy as np
ts = pd.Series(np.random.randint(0,10,len(rng)), index=rng)
ts.head(20)
```
