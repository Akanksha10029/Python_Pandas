```python
#Pivot basics
import pandas as pd
import numpy as np
df = pd.read_csv("/content/pivot_Table_weather.csv")
df
```

```python
df.pivot(index='city',columns='date')
```

```python
df.pivot(index='city',columns='date',values="temperature")
```

```python
df.pivot(index='city',columns='date',values="humidity")
```

```python
df.pivot(index='date',columns='city')
```

```python
#Pivot Table
df=pd.read_csv('/content/pivot_Table_weather2.csv')
df
```

```python
'''In this data set we are having diffrent values of temp. and humidity on the same date like for morning and night
so when we organise the date in the form of table like the rows are city and
columns are date then it will give the average of the values on the same date of that particular city'''
df.pivot_table(index="city",columns="date")
```

```python
#Margins
'''by default margins is false but here we have done it true so it has given
the sum of all the rows and columns in All column and 1 All row at the end.
here aggfunc is aggregate function which use numpy for the maths calculation like sum, mean, avg etc.'''
df.pivot_table(index="city",columns="date", margins=True, aggfunc=np.sum)
```
