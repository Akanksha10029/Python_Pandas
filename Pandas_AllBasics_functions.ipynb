```python
'''describe() provides summary statistics of the numeric columns within that DataFrame.
Count: The number of non-null values in each numeric column.
Mean: The mean (average) value of each numeric column.
Standard Deviation (std): The standard deviation, which measures the dispersion or spread of data, for each numeric column.
Minimum (min): The minimum value in each numeric column.
25th Percentile (25%): The value below which 25% of the data falls for each numeric column (the first quartile).
50th Percentile (50%): The median or middle value for each numeric column (the second quartile).
75th Percentile (75%): The value below which 75% of the data falls for each numeric column (the third quartile).
Maximum (max): The maximum value in each numeric column.
'''
df.describe()
```
