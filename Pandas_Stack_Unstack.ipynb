```python
#Reshape dataframe using stack/unstack

import pandas as pd
df=pd.read_excel('/content/stacks_stocks_2_levels.xlsx',header=[0,1]) #Here header[0,1] means 2 header level columns in the top
df
```

```python
'''by default by calling stack(),i.e, without any argument then it transpose the inner most level
,i.e,Facebook, google, microsoft.....'''
df.stack()
```

```python
df.stack(level=0)
```

```python
df.stack(level=1)
```
