```python
import pandas as pd
df=pd.read_csv('/content/weather_data.csv', parse_dates=['day'])
df
```

```python
new_df=df.fillna(
    {
        'temperature':0,
        'windspeed':0,
        'event':'no event'
    }
)
new_df
```

```python
new_df=df.fillna(method='ffill',limit=1)  #ffill=forward fill (in this previous values of rows is being carryforward to the null-space)
new_df              #In 'limit'we have specified a limit that to how rows we want to carry forward the value of prvious one
```

```python
new_df=df.fillna(method='bfill')#bfill=backward fill in which the downward value which comes first towards reaching null-space,then it carry forward the value of downward values to upwards
new_df
```

```python
new_df=df.interpolate(method='time')
new_df
```

```python
''' thresh=2 drops that particular row in which atleast 2 valid values are there...
eg here in row 6 there was only one valid value i.e 	2017-01-09
 and there is no other valid value or there is only NaN '''
new_df=df.dropna(thresh=2)
new_df
```
