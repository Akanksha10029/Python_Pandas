```python
#Crosstab in pandas
import pandas as pd
df=pd.read_excel("/content/Crosstab_survey.xls")
df
```

```python
'''Here we are arranging rows and columns according to our requirements.
df.Nationality=row or x-axis' and df.handedness=column or y-axis'''
pd.crosstab(df.Nationality,df.Handedness)
```

```python
pd.crosstab(df.Sex, df.Age)
```

```python
#Margins
pd.crosstab(df.Nationality, df.Name, margins=True) #margins=total(i.e All)
```

```python
# Multi Index Column and Rows
'''Here df.Sex=x-axis i.e rows' and [df.Handedness,df.Nationality]=y-axis i.e columns.
handedness will be at top as it is written at first and Nationality after this'''
pd.crosstab(df.Sex, [df.Handedness,df.Nationality], margins=True)
```

```python
pd.crosstab([df.Nationality, df.Sex],[df.Handedness],margins=True)
```

```python
# Normalise
pd.crosstab(df.Sex, df.Handedness, normalize='index') # normalise gives the percentage
```

```python
# Aggfunc and Values
'''Here we are finding average on the basis of ages that's why taken the values as Age and
by using aggfunc we found average with the hepl of numpy'''
import numpy as np
pd.crosstab(df.Sex, df.Handedness, values=df.Age, aggfunc=np.average)
```
